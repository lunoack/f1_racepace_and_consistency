```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
def setup_fastf1():
    fastf1.Cache.enable_cache('../fantasy_cache')
```

```python
def session_data(year, gp, session):
    session = fastf1.get_session(year, gp, session)
    session.load()
    data_base = session.laps.pick_accurate().reset_index()
    return data_base
```

```python
if __name__ == "__main__":
    setup_fastf1()

    year = 2025
    today = pd.Timestamp(datetime.now().date())

    schedule = fastf1.get_event_schedule(year)

    past_races = schedule[
        (schedule['EventDate'] < today) &
        (schedule['EventFormat'] != 'testing')
    ].sort_values('EventDate')


    exit()

    last_5_races = past_races.tail(5)['EventName'].tolist()

    data_base = pd.DataFrame()

    for gp in last_5_races:
        for sess in [1, 2, 3, 'Q', 'R']:
            try:
                sess_data = session_data(year, gp, sess)
                data_base = pd.concat([data_base, sess_data], ignore_index=True)
            except Exception:
                continue
```
